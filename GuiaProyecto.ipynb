### 1.3 Definición del problema.
> Se debe describir en una cuartilla máximo cuál es el escenario de negocio que se va a modelar, identificar cuáles son los supuestos que defines en el negocio y cuáles son los "nodos" o "escenarios" que identificas necesarios simular y porqué. Debes elaborar un diagrama de inicio a fin, identificando los "nodos" o "escenarios" a simular.



Para este proyecto se van a modelar 4 distintos tipos de cambio, los cuáles son el dólar (USD/MXN), euro (EUR/MXN), yuan (CNY/MXN) y el yen (JPY/MXN). Como se puede observar todas las divisas con las que trabajaremos serán respecto al peso mexicano. Para poder modelar obtuvimos los datos mensuales de los tipos de cambio de cada una de las divisas mencionadas desde el mes de abril del 2015 hasta el mes de marzo del 2023. 

A continuación, se muestra el diagrama para realizar nuestro modelo donde se identifican los “nodos” o “escenarios” a simular.


Los nodos que tomamos en cuenta en este proyecto son los que creemos que pueden tener un impacto y a su vez guardan una relación con el tipo de cambio. 
El diferencial de la inflación se refiere a la diferencia que existe entre la tasa de inflación de dos o más países, se va a utilizar como “nodo” ya que suele utilizarse como indicador de la posición competitiva y de la variación relativa de los tipos de cambio a largo plazo.



En nuestro modelo asumimos que el tipo de cambio entre dos monedas pueden predecirse con estos nodos.
