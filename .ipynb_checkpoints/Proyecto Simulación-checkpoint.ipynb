# Proyecto TEMA-2

## Definición del proyecto

Para este proyecto se van a modelar 4 distintos tipos de cambio, los cuáles son el **dólar (USD/MXN)**, **euro (EUR/MXN)**, **yuan (CNY/MXN)** y el **yen (JPY/MXN)**. Como se puede observar todas las divisas con las que trabajaremos serán respecto al peso mexicano. Para poder modelar obtuvimos los datos mensuales de los tipos de cambio de cada una de las divisas mencionadas desde el mes de abril del 2015 hasta el mes de marzo del 2023. 

A continuación, se muestra el diagrama para realizar nuestro modelo donde se identifican los “nodos” o “escenarios” a simular.

En nuestro modelo tenemos los siguientes dos supuestos:
1. Asumimos que todas las actividades son estadísticamente independientes 
2. Asumimos que el tipo de cambio entre dos monedas pueden predecirse con estas variables.

Las variables que tomamos en cuenta en este proyecto son los que creemos que pueden tener un impacto y a su vez guardan una relación con el tipo de cambio. 

- El diferencial de la inflación se refiere a la diferencia que existe entre la tasa de inflación de dos o más países, se va a utilizar como variable ya que suele utilizarse como indicador de la posición competitiva y de la variación relativa de los tipos de cambio a largo plazo.
- El diferencial de crecimiento económico hace referencia a la diferencia que existe entre la tasa de crecimiento económico entre dos o más países, se va a utilizar como variable ya que hay una relación en el largo plazo entre el aumento o dismunición del PIB con el tipo de cambio, por ejemplo un crecimiento del producto interno bruto aprecia la moneda de ese país, es decir, el incremento de la productividad de un país en relación con otros hace que su moneda se aprecie.
- Por otra parte la tasa de interés entre ambos países nos habla acerca de la tasa de interés interna que maneja cada país en el que se ve involucrado el tipo de cambio y hay una relación a corto plazo entre tipo de cambio y tasa de interés por ejemplo si la tasa de interés interna sube y se mantiene constante el tipo de cambio, el rendimiento de los activos locales se eleva en relación con los extranjeros por lo que los inversionistas los demandarán más, ocasionando que la moneda local se aprecie y vicervesa.
- Pasando a las expectativas del crecimiento económico estas son un determinante  importante para el tipo de cambio. Es decir, el valor de la moneda local, depende de su valor de reventa futuro. En contraste, los factores que provoquen una caída en el valor esperado del tipo de cambio en el futuro, forzarán sobre la depreciación del tipo de cambio en el periodo actual.
- Por último tenemos las barreras arancelarias, las cuáles son una medida o restricción que impone el Gobierno de un país. Esto, para reducir o eliminar el intercambio de bienes y/o servicios con otra(s) nación(es). Se va a utilizar como variable ya que se considera un determinante en el corto plazo, por ejemplo la reducción de aranceles va acompañada de una depreciación inmediata del tipo de cambio real.
